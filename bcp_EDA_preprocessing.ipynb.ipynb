### Visualization of selected features against target column.
The target column here is "diagnosis". Important fetures are selected and visualized against target columns.

```python
featurecols = ["diagnosis", "radius_mean", "texture_mean", "perimeter_mean", "area_mean"]

sns.pairplot(bcpdata[featurecols], hue="diagnosis")
plt.show()
```

### **Analysis of the Pairplot**

#### **1. Data Distribution**
- The diagonal elements show kernel density estimation (KDE) plots, displaying the distribution of each feature.
- There is a distinction between the two diagnosis categories, "M" (Malignant) and "B" (Benign), represented in different colors (blue and orange).

#### **2. Feature Relationships**
- **Highly Correlated Features:**  
  - The features **radius_mean, perimeter_mean, and area_mean** are strongly correlated, as seen from their near-linear relationships in the scatterplots.
- **Texture_mean:**  
  - Shows more spread, with less apparent correlation with other features.
- **Separation of Malignant and Benign Cases:**  
  - The malignant cases (blue) tend to have higher values for **radius_mean, perimeter_mean, and area_mean**, while benign cases (orange) cluster in the lower range.

#### **3. Possible Insights**
- **Predictive Power:** Features such as **radius_mean, perimeter_mean, and area_mean** might be strong predictors for distinguishing between benign and malignant cases.
- **Multicollinearity Issue:** The strong correlations indicate potential redundancy, which could impact machine learning models like logistic regression.

### Creating Scatter plots

This code creates a bubble plot of texture_mean vs. radius_mean, where each point has a random size and color. It is useful for exploratory visualization but should be refined for clearer insights.

```python
size = len(bcpdata['texture_mean'])

area = np.pi * (15 * np.random.rand( size ))**2
colors = np.random.rand( size )

plt.xlabel("texture mean")
plt.ylabel("radius mean") 
plt.scatter(bcpdata['texture_mean'], bcpdata['radius_mean'], s=area, c=colors, alpha=0.5);
```

This visualization effectively compares tumor texture and radius, while incorporating tumor area as size and color. It’s a great exploratory data analysis (EDA) tool for understanding breast cancer features.

```python
# Define size as the number of data points
size = len(bcpdata['texture_mean'])

# Scale the area sizes based on 'area_mean' feature
area = np.pi * (bcpdata['area_mean'] / bcpdata['area_mean'].max())**2 * 100  # Normalized and scaled

# Set colors based on another feature (optional)
colors = bcpdata['area_mean'] / bcpdata['area_mean'].max()  # Normalized colors

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(bcpdata['texture_mean'], bcpdata['radius_mean'], s=area, c=colors, cmap='viridis', alpha=0.5)

# Labels and title
plt.xlabel("Texture Mean")
plt.ylabel("Radius Mean")
plt.colorbar(label="Normalized Area Mean")  # Add a color legend
plt.title("Scatter Plot of Texture Mean vs Radius Mean with Scaled Marker Sizes")

plt.show()
```

### Encoding the catagorical data in the target column


```python
from sklearn.preprocessing import LabelEncoder
```

```python
bcpdata.head(2)
```

The code uses LabelEncoder() to convert the categorical diagnosis column (M for Malignant, B for Benign) into numerical values (1 and 0, respectively) for machine learning models, as they require numerical input. This ensures proper processing but may not be ideal for multi-class categories.

```python
labelencoder_Y = LabelEncoder()
bcpdata.diagnosis = labelencoder_Y.fit_transform(bcpdata.diagnosis)
```

### Generating correlation matrix of selected columns.
The code selects 11 columns, including diagnosis and tumor-related features, then computes their correlation matrix using .corr(), which measures the linear relationships between variables. 

```python
cols = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']
print(len(cols))
bcpdata[cols].corr()
```

```python
plt.figure(figsize=(12, 9))

plt.title("Correlation Graph")

cmap = sns.diverging_palette( 1000, 120, as_cmap=True)
sns.heatmap(bcpdata[cols].corr(), annot=True, fmt='.1%',  linewidths=.05, cmap=cmap);
```

Using, Plotly Package we can show it in interactive graphs like this,

```python
plt.figure(figsize=(15, 10))


fig = px.imshow(bcpdata[cols].corr());
fig.show()
#since it is an interactive plot, you can hover over the cells to see the correlation values. That is why it is not visible here, but will be visible in the notebook.
```
